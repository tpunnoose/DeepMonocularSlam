```python
import odometry

import numpy as np
import cv2 as cv
import pykitti

import matplotlib.pyplot as plt
```

```python
basedir = '/Users/tarun/Classes/CS231A/project/KITTI/odometry/dataset'
sequence = '04'
dataset = pykitti.odometry(basedir, sequence)

so = odometry.StereoOdometry(dataset)

so.process_next_img()
so.process_next_img()
```

```python
img_prev = so.prev_stereo_frame.left.image
img_next = so.next_stereo_frame.left.image

kp_prev = so.prev_stereo_frame.left.keypoints
kp_next = so.next_stereo_frame.left.keypoints

so.matches = sorted(so.matches, key = lambda x:x.distance)

image_matches = cv.drawMatches(img_prev,kp_prev,img_next,kp_next,so.matches[:50],None,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

plt.imshow(image_matches),plt.show()

cv.imwrite("matched_image.png", image_matches)
```

```python
image_keypoints = cv.drawKeypoints(img_prev,kp_prev,None, flags=0)

plt.imshow(image_keypoints),plt.show()

cv.imwrite("image_keypoints.png", image_keypoints)
```
